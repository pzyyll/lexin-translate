```python
import deepl
import dotenv
import os

dotenv.load_dotenv()

deepl_auth_key = os.getenv('DEEPL_AUTH_KEY')
client = deepl.Translator(deepl_auth_key)

result = client.translate_text('Hello, World!', target_lang='ZH')

print(result.text)
```

```python
from pydantic import BaseModel, RootModel
from typing import List

class Language(BaseModel):
    language: str
    name: str

Languages = RootModel[List[Language]]

lang = Language.model_validate_json('{"language": "EN", "name": "English"}')
langs = Languages.model_validate_json('[{"language": "EN", "name": "English"}]')
print(lang)
print(langs)
```

```python
import httpx

url = "https://api-free.deepl.com/v2"
auth_key = os.getenv("DEEPL_AUTH_KEY")
headers = {
    "Authorization": f"DeepL-Auth-Key {auth_key}",
    "Content-Type": "application/json",
}

async with httpx.AsyncClient() as client:
    # Get languages
    response = await client.post(
        f"{url}/languages",
        headers=headers,
    )
    
    languages = Languages.model_validate_json(response.text)
    
    # Get language names and translate them
    names = [language.name for language in languages.root]
    
    response = await client.post(
        f"{url}/translate",
        headers=headers,
        json={
            "text": names,
            "target_lang": "ZH",
        },
    )
    
    # Update language names with translations
    translations = response.json()["translations"]
    
    for i, language in enumerate(languages.root):
        language.name = translations[i]["text"]

    # dumps json file
    with open("languages.json", "w") as f:
        f.write(languages.model_dump_json(indent=2))
```

```python
print(response.json())

for translation in response.json()["translations"]:
    print(translation["text"])
```
